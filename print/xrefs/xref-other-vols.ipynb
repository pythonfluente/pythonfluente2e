```python
from pathlib import Path
base_online = Path('../../online/')
base_vol = Path('../../vol2/')
cap_n = 9
cap = f'cap{cap_n:02d}.adoc'

print(base_vol / cap)
```

```python
def readlines(base, cap_n):
    cap = f'cap{cap_n:02d}.adoc'
    with open(base / cap) as fp:
        return fp.readlines()
```

```python
other_vols = {}

for cap_n in list(range(1, 9)) + list(range(17, 25)):
    lines = readlines(base_online, cap_n)
    for line in lines:
        if line.startswith('[[ch_'):
            ident = line.split(']]')[0][2:]
            other_vols[ident] = (((cap_n - 1) // 8) + 1, cap_n)

other_vols
```

```python
# duas xrefs:
# A maior parte do <<ch_type_hints_def>> e do <<ch_more_types>>

for cap_n in range(13, 14):
    lines = readlines(base_online, cap_n)
    for line in lines:
        if 'ch_type_hints_def' in line:
            print(line)
        for ident, (vol, cap) in other_vols.items():
            xref = f'<<{ident}>>'
            new = ''
            if xref in line:
                print(line, end='')
                new = line.replace(xref, f'https://fpy.li/{cap}[«Capítulo {cap}»] (vol.{vol})')
                print(new, end='')
            line = new
        assert '<<ch' not in line, line
```
