```python
def gen_sample(encoding: str) -> list[str]:
    res = []
    for i in range(128, 256):
        c = bytes([i]).decode(encoding, errors='replace')
        res.append(c)
    return res
```

```python
ibm = set(gen_sample('cp437'))
ms = set(gen_sample('cp1252'))
```

```python
for c in sorted(ibm & ms):
    print(c, end=' ')
```

```python
for c in sorted(ibm - ms):
    print(c, end=' ')
```

```python
for c in sorted(ms - ibm):
    print(c, end=' ')
```
